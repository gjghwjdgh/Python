```python
t = np.arange(0,12,1)
y1 = np.sin(t)
y2 = np.cos(t)
```

```python
plt.plot(y1,y2,'--x',color="red")
```

```python
plt.plot(t,y1,'r--',marker="*")
plt.plot(t,y2,'go-')
plt.grid()
plt.figure(figsize = (12,2))
```

```python
plt.subplot(1,2,1)
plt.plot(t,y1,'r--',marker="*")
plt.subplot(1,2,2)
plt.plot(t,y2,'go-')
```

- supplot 을 이용해서 2행 1열 구조 2개씩
- 4가지 색상 다르게
- 4가지 스타일 다르게
- 4가지 마커 서로 다르게
- 범례표시
- x,y,제목

# x = 0부터 5까지 단위1

# y1=x+1

# y2=x+2

# y3=x+3

# y4=x+4

```python
x = np.arange(0,5,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4
plt.subplot(1,2,1)
plt.xlabel("universe")
plt.ylabel("star")
plt.title("space")
plt.plot(x,y1,'-*',color = "cyan")
plt.plot(x,y2,'--+',color = "k")
plt.legend("st")

plt.subplot(1,2,2)
plt.xlabel("molecule")
plt.ylabel("atom")
plt.title("particle")
plt.plot(x,y3,'--s',color = "skyblue")
plt.plot(x,y4,'-p',color = "pink")
plt.legend("as")
```
