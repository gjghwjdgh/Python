```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
path = r"C:\Users\user\Desktop\python\h.CSV"
```

```python
path
```

```python
df=pd.read_csv(path, encoding='cp949')
df
t = df
```

```python
a=t["이름"]
b=t["국어"]
c=t["수학"]
```

```python
a,b,c
```

```python
rc('font',family = "gulim")
plt.subplot(1,2,1)
plt.plot(a,b,"--o",color = "navy")
plt.ylabel('국어점수')
plt.xlabel('이름')
plt.title("국어점수")
plt.legend("국어")
plt.subplot(1,2,2)
plt.plot(a,c,"--x",color = "blue")
plt.ylabel('수학점수')
plt.xlabel('이름')
plt.title("수학점수")
plt.legend("수학")
plt.show()
```
