```python
import pandas as pd
```

```python
path = 'data.csv'
```

```python
df = pd.read_csv(path,encoding='cp949')
```

```python
df.head()
```

```python
null_count = df.isnull().sum()
```

```python
df_null_count = null_count.reset_index()
```

```python
df_null_count.columns=['메뉴','결측']
```

```python
df_null_count_3 = df_null_count.sort_values(by='결측',ascending=False).head(11)
df_null_count_3
```

```python
drop_columns = df_null_count_3['메뉴']
```

```python
drop_columns.to_list()
```
